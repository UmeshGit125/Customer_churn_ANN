```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler , LabelEncoder   
import pickle  

  
```

```python
data = pd.read_csv('Churn_Modelling.csv')  
data
```

```python
data = data.drop(['RowNumber', 'CustomerId' , 'Surname'] ,axis=1)     
data.head()
```

```python
# encoding categorical variable 
label_gender_encoder = LabelEncoder() 
data['Gender'] = label_gender_encoder.fit_transform(data['Gender'])   
data
```

```python
# Geopgraphy encoding using one hot encoding 
    
from sklearn.preprocessing import OneHotEncoder   
onehot_encoder_geo=OneHotEncoder( ) 
geo_encoder= onehot_encoder_geo.fit_transform(data[['Geography']]).toarray()
geo_encoder
```

```python
 
onehot_encoder_geo.get_feature_names_out(['Geography'])  
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns = onehot_encoder_geo.get_feature_names_out(['Geography']))  
geo_encoded_df
```

```python
data
```

```python
data = pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis = 1 )  
```

```python
# storing the pickle file 
with open('label_gender_encoder.pkl','wb') as file: 
    pickle.dump(label_gender_encoder,file)
```

```python
with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
 

    
```

```python
# Divide data in dependend and independend variables 
x = data.drop('Exited',axis=1)
y = data['Exited']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)


#scaling these feature 

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# saving scaler file 
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler, file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime 
import keras 
```

```python
data
```

```python
x_train.shape[1]
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape = (x_train.shape[1],)), # input layer  
    Dense(32,activation='relu'), # hidden layer 
    Dense(1,activation='sigmoid')   # output layer 
 ]
    
)
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.BinaryCrossentropy()
loss 
```

```python
model.compile(optimizer=optimizer,loss=loss,metrics=['accuracy'])
```

```python
## Set up the tensorboard 
import tensorflow  
# import tensoflow.keras.callbacks 
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping 
 
log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") 
tensorflow_callback = TensorBoard(log_dir= log_dir , histogram_freq= 1 )
```

```python
# seeting up the early stopping 

early_stopping_callback =EarlyStopping(monitor='val_loss' , patience= 10 , restore_best_weights= True)
```

```python
## Train model 
history = model.fit(x_train, y_train , 
                    validation_data= (x_test , y_test),
                    epochs=100,
                    callbacks = [tensorflow_callback , early_stopping_callback  ]
                    )
```

```python
model.save('model.h5')
```
